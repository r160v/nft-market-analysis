# Number of NFTs per category

```python
targetdir = "./data/NFTsPerCategory/"
filename = "number_of_nft_per_category.csv"

number_of_nft_per_category = pd.read_csv(targetdir + filename)
```

```python
number_of_nft_per_category["es_cat"] = ["Utilidades", "Metaverso", "Otros", "Arte", "Coleccionables", "Juegos"]
number_of_nft_per_category
```

```python
def my_autopct(pct):
    return ('%.2f' % pct) if pct > 20 else ''


colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99', '#88aa99', '#eb3462']
explode = (0.2,0.12,0.05,0.05,0.05,0.05)
 
fig1, ax1 = plt.subplots()
ax1.pie(number_of_nft_per_category["distinct_nft"], colors = colors, explode=explode,labels=number_of_nft_per_category["es_cat"], 
        autopct='%1.1f%%', startangle=90, pctdistance=0.85, textprops={'fontsize': 11})  #draw circle

centre_circle = plt.Circle((0,0),0.70,fc='white')
matplotlib.rcParams.update({'font.size': 8})
plt.rcParams["figure.figsize"] = (10,5)

fig = plt.gcf()
fig.gca().add_artist(centre_circle) # Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis('equal')  
plt.tight_layout()

plt.savefig("./nftPerCategory/nft_per_category.svg", bbox_inches='tight')
plt.show()
```

# Number of NFT per collection

```python
targetdir = "./data/NFTsPerCollection/"
filename = "number_of_nft_per_collection.csv"

number_of_nft_per_collection = pd.read_csv(targetdir + filename)
```

```python
number_of_nft_per_collection.sort_values(['category', 'distinct_nft'],
              ascending = [True, False], inplace=True)

number_of_nft_per_collection
```

```python
def process_df(df, category):
    x = df[df["category"] == category][:5].reset_index()
    x.drop(["index"],  axis=1, inplace=True)
    x["distinct_nft"] = x["distinct_nft"] / x["distinct_nft"].max()
    
    return x
```

```python
number_of_nft_per_collection[number_of_nft_per_collection["category"] == "Art"].sort_values(["distinct_nft"], 
                                                                                            ascending=False)[:10]
```

```python
x = number_of_nft_per_collection[number_of_nft_per_collection["category"] == "Art"][:5].reset_index()
x
```

```python
x["percentage"] = x["distinct_nft"] / x["distinct_nft"].sum()
x
```

```python
art_df = process_df(number_of_nft_per_collection, "Art")
art_df[:20]
```

```python
collectible_df = process_df(number_of_nft_per_collection, "Collectible")
collectible_df
```

```python
games_df = process_df(number_of_nft_per_collection, "Games")
games_df
```

```python
metaverse_df = process_df(number_of_nft_per_collection, "Metaverse")
metaverse_df
```

```python
other_df = process_df(number_of_nft_per_collection, "Other")
other_df
```

```python
utility_df = process_df(number_of_nft_per_collection, "Utility")
utility_df
```

```python
fig, ax = plt.subplots()
# plt.rcParams["figure.figsize"] = (10,10) # OK
plt.rcParams["figure.figsize"] = (15,10)
ax.set_xlim(-1, 5)
ax.set_ylim(6, 10.5)
plt.style.use('_classic_test_patch')

# Art
ax.scatter(["1", "2", "3", "4", "5"], [10, 10, 10, 10, 10], 
                  color='darkblue', 
                 alpha=0.7,
                 s = art_df["distinct_nft"] * 2000, label='Arte')

for i, txt in enumerate(art_df["collection"]):
    ax.annotate(txt, (["1", "2", "3", "4", "5"][i], [10.25, 10.25, 10.25, 10.25, 10.25][i]), size=16, ha='center')


# Collectible
ax.scatter(["1", "2", "3", "4", "5"], [9.25, 9.25, 9.25, 9.25, 9.25], color='purple', 
               alpha=0.7,
                 s = collectible_df["distinct_nft"] * 2000, label='Coleccionables')

for i, txt in enumerate(collectible_df["collection"]):
    ax.annotate(txt, (["1", "2", "3", "4", "5"][i], [9.5, 9.5, 9.5, 9.5, 9.5][i]), size=16, ha='center')
    
    
# Games
ax.scatter(["1", "2", "3", "4", "5"], [8.5, 8.5, 8.5, 8.5, 8.5], color='green', 
               alpha=0.7,
                 s = games_df["distinct_nft"] * 2000, label='Juegos')

for i, txt in enumerate(games_df["collection"]):
    ax.annotate(txt, (["1", "2", "3", "4", "5"][i], [8.75, 8.75, 8.75, 8.75, 8.75][i]), size=16, ha='center')
    

# Metaverse
ax.scatter(["1", "2", "3", "4", "5"], [7.75, 7.75, 7.75, 7.75, 7.75], color='red', 
               alpha=0.7,
                 s = metaverse_df["distinct_nft"] * 2000, label='Metaverso')

for i, txt in enumerate(metaverse_df["collection"]):
    ax.annotate(txt, (["1", "2", "3", "4", "5"][i], [8, 8, 8, 8, 8][i]), size=16, ha='center')
    

# Other
ax.scatter(["1", "2", "3", "4", "5"], [7, 7, 7, 7, 7], color='yellow', 
               alpha=0.7,
                 s = other_df["distinct_nft"] * 2000, label='Otros')

for i, txt in enumerate(other_df["collection"]):
    ax.annotate(txt, (["1", "2", "3", "4", "5"][i], [7.25, 7.25, 7.25, 7.25, 7.25][i]), size=16, ha='center')
    
    
# Utility
ax.scatter(["1", "2", "3", "4", "5"], [6.25, 6.25, 6.25, 6.25, 6.25], color='black', 
               alpha=0.7,
                 s = utility_df["distinct_nft"] * 2000, label='Utilidades')

for i, txt in enumerate(utility_df["collection"]):
    ax.annotate(txt, (["1", "2", "3", "4", "5"][i], [6.5, 6.5, 6.5, 6.5, 6.5][i]), size=16, ha='center')
    
lgnd = ax.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc='lower left',
                      ncol=6, mode="expand", borderaxespad=0.5, prop={'size': 16}, markerscale=.3, frameon=True)

lgnd.get_frame().set_edgecolor("black")
plt.axis('off')

plt.savefig("./NFTPerCollection/top_five_collections_per_category.svg", bbox_inches='tight')
```
